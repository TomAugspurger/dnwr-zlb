# Earnings by Demographics

```python
df = earn_store.select('m1996_12')  # first valid
df = get_useful(df)
df = df.replace(-1, np.nan)
df = replace_categorical(df)
ids = df.xs(4, level='minor')
earnings = df.earnings.unstack()
diff = earnings[8] - earnings[4]
```

```python
# Race:
print(earnings.groupby(ids['race']).median())
print(earnings.groupby(ids['race']).size())
r = earnings.groupby(ids['race']).median().mean(1)
r.sort()
r.plot(kind='barh')
```

```python
earnings.groupby(ids['sex']).mean()
```

```python
r = earnings.groupby(ids['age'].dropna().astype(np.int64)).median()
r.mean(1).plot()
pd.rolling_mean(r.mean(1), 4).plot()
```

```python
r = earnings.groupby(ids['married']).median()
r
r.mean(1).plot(kind='barh')
```

```python
r = earnings.groupby([ids['sex'], ids['married']]).median()
r.mean(1).unstack(0).plot(kind='barh')
```

```python
r = earnings.groupby(ids['labor_status']).median()
r.mean(1)
```

```python
r = diff.groupby(ids['labor_status']).median().copy()
r.sort()
print(r)
r.plot(kind='barh')
```
