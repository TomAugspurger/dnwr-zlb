# Let's try a Regression

```python
df = get_useful(earn_store.select('m1996_12'))
df = replace_categorical(df)
e = df.earnings.unstack().dropna()
e['male'] = df.sex.unstack()[4] == 'male'
e['white'] = df.race.unstack()[4] == 'White Only'
e['edu'] = df.edu.unstack()[4]
e = e[(e[4] > 0) & (e[8] > 0)]
e['ln_wage_change'] = np.log(e[8]) - np.log(e[4])
```

```python
mod = sm.OLS.from_formula('ln_wage_change ~ edu + male + white', e)
res = mod.fit()
res.summary()
```

# "Grouping" by employment history

```python
keys = panel_store.keys()
m0 = arrow.get(keys[0], '/mYY-MM')
mn = arrow.get(keys[-1], '/mYY_MM')
months = (x.strftime('%Y_%m') for x in arrow.Arrow.range('month', m0, mn))
months = (month for month in months if '/m' +  month in keys)
```

```python
g4s = []
g8s = []
for month in months:
    wp = get_useful(panel_store.select('m' + month))
    e_types = ['either', 'unemployed', 'nonemployed']
    try:
        [add_employedment_status_last_period(wp, kind=x) for x in e_types]
    except KeyError:
        continue
    stamps = [pd.Timestamp(pd.datetime(int(wp['year'][m].dropna().values[0]),
                                       int(wp['month'][m].dropna().values[0]), 1))
              for m in (4, 8)]

    race = replace_categorical(wp['race'], kind='race')
    sex = replace_categorical(wp['sex'], kind='sex')
    age = wp['age']
    
    group4 = [[race[4], sex[4], age[4], wp[x + '_last_period'][4]] for x in e_types]
    group8 = [[race[m], sex[8], age[8], wp[x + '_last_period'][8]] for x in e_types]
    
    def make_group(m, groups, stamp):
        r = pd.concat([wp['earnings'][m].groupby(group).median()
                      for group in groups], keys=e_types).unstack()
        r.index.names = ['employment', 'race', 'sex', 'age']  # set_names?
        r['stamp'] = stamp
        r = r.set_index('stamp', append=True)
        r = r.reorder_levels(['stamp', 'employment', 'sex', 'race', 'age'])
        return r

    r4 = make_group(4, group4, stamps[0])
    r8 = make_group(8, group8, stamps[1])
    
    g4s.append(r4)
    g8s.append(r8)
```

```python
df = pd.concat(g4s).sort_index()
```
